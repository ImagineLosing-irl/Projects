# Identifying and Removing the outliers

```python
sns.boxplot(car_data['price'])
```

```python
sorted(car_data['price'],reverse=True)
```

```python
car_data = car_data[~((car_data['price'] >= 40960.0) & (car_data['price'] <= 45400.0))]
```

# Normalization

```python
from sklearn.preprocessing import MinMaxScaler
numeric_cols = car_data.select_dtypes(include = [np.number])
non_numeric_cols = car_data.select_dtypes(exclude = [np.number])
scaler = MinMaxScaler()
scaled_numeric_data = scaler.fit_transform(numeric_cols)
scaled_numeric_df = pd.DataFrame(scaled_numeric_data, columns = numeric_cols.columns)
scaled_data = pd.concat([scaled_numeric_df, non_numeric_cols.reset_index(drop = True)], axis = 1)
print(scaled_data.shape)
print()
print('*' * 60)
scaled_data.head()
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
numeric_cols = car_data.select_dtypes(include = [np.number])
non_numeric_cols = car_data.select_dtypes(exclude = [np.number])
scaler = StandardScaler()
scaled_numeric_data = scaler.fit_transform(numeric_cols)
scaled_numeric_df = pd.DataFrame(scaled_numeric_data, columns = numeric_cols.columns)
scaled_data = pd.concat([scaled_numeric_df, non_numeric_cols.reset_index(drop = True)], axis = 1)
print(scaled_data.shape)
print()
print('*' * 60)
scaled_data.head()
```
